```python
def outliersFunction(df,atr):
    arr1 = df[atr].values.astype(float)
    q1 = np.quantile(arr1, 0.25)
    q3 = np.quantile(arr1, 0.75)
    med = np.median(arr1)
    iqr = q3-q1
    upper_bound = q3+(1.5*iqr)
    lower_bound = q1-(1.5*iqr)

    outliers = arr1[(arr1 <= lower_bound) | (arr1 >= upper_bound)]
    print('The following are the outliers in {} : {}'.format(atr,outliers))
    print('Total number of outliers: ',outliers.size)
    porcentaje = outliers.size*100/bd_ipsa[atr].count()
    print('Percentage of outliers: ',outliers.size*100/df[atr].count())
    return outliers, porcentaje
```

```python
for n in ['edad', 'lluvias', 'me', 'pct_diatrea', 'sacarosa', 'semsmad', 'vejez']:
    outliersFunction(bd_ipsa, n)
```

```python
numeric_cols = ['edad', 'lluvias', 'me', 'pct_diatrea', 'sacarosa', 'semsmad', 'vejez']
for col in numeric_cols:
  ouliers, porcentaje = outliersFunction(bd_ipsa,col)
  if porcentaje <= 10:
    bd_ipsa = bd_ipsa.loc[~bd_ipsa[col].isin(ouliers)]
    bd_ipsa = bd_ipsa.reset_index(drop=True)
```
